```python
import random
import string


# INPUT PHASE
def get_password_len():

    while True:

        user_input = input("Enter Your Password Length (8 to 64): ")

        if user_input.isdigit():

            length = int(user_input)

            if 8 <= length <= 64:

                return length

            else:
                print("Error: Password length must be between 8 and 64")

        else:
            print("Please Enter a Valid Number")


# PROCESS PHASE
def generate_password(length):

    all_char = (
        string.ascii_letters +
        string.digits +
        string.punctuation
    )

    password = "".join(
        random.choice(all_char)
        for _ in range(length)
    )

    return password


# OUTPUT PHASE
def display_password(password):

    print(f"\nYour New Password Is: {password}")


# MAIN FUNCTION
def main():

    length = get_password_len()

    password = generate_password(length)

    display_password(password)


# PROGRAM START
if __name__ == '__main__':

    main()
```
