```python
# ============ To Do List Program ============

File_Name = 'tasks.txt'

# List to store tasks
my_tasks = []

# Load old tasks from file
try:
    with open(File_Name, 'r') as f:
        lines = f.readlines()

        for i, line in enumerate(lines, start=1):
            my_tasks.append({
                'id': i,
                'Task': line.strip()
            })

except FileNotFoundError:
    pass


# Function to display tasks
def display_task():

    if not my_tasks:
        print("\nYour Tasks Are Not Found\n")

    else:
        print("\nYour Tasks Here")
        print("=" * 40)

        for t in my_tasks:
            print(f"{t['id']} : {t['Task']}")

        print("=" * 40)


# Function to add a task
def add_task():

    task = input("Enter your task: ")

    task_id = len(my_tasks) + 1

    task_dict = {
        'id': task_id,
        'Task': task
    }

    my_tasks.append(task_dict)

    # Save task in file
    with open(File_Name, 'a') as file:
        file.write(task + '\n')

    print(f"Task '{task}' added successfully!\n")


# Main Program
def main():

    print("=" * 50)
    print("Welcome To Do List Program")
    print("=" * 50)

    while True:

        print("\nOptions")
        print("1 : Enter Your Task")
        print("2 : See Your Tasks")
        print("3 : Exit Program")

        choice = input("Enter Your Choice: ")

        if choice == '1':
            add_task()

        elif choice == '2':
            display_task()

        elif choice == '3':
            print("Program Exited Successfully")
            break

        else:
            print("Invalid Choice! Please enter 1 to 3")


# Program Start
if __name__ == '__main__':
    main()
```
