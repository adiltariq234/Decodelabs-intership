```python
expenses = []
incomes  = []


def add_expense():
    print("\n--- Expense Add Karo ---")
    date     = input("Tarikh (jaise 2024-01-25): ")
    category = input("Category (Food/Transport/Shopping/Other): ")
    amount   = float(input("Amount: "))
    note     = input("Note (optional): ")

    expenses.append({
        "date": date,
        "category": category,
        "amount": amount,
        "note": note
    })
    print("Expense save ho gaya!")


def add_income():
    print("\n--- Income Add Karo ---")
    date   = input("Tarikh (jaise 2024-01-25): ")
    source = input("Source (Salary/Freelance/Other): ")
    amount = float(input("Amount: "))
    note   = input("Note (optional): ")

    incomes.append({
        "date": date,
        "source": source,
        "amount": amount,
        "note": note
    })
    print("Income save ho gayi!")


def show_expenses():
    print("\n--- Apke Expenses ---")
    if not expenses:
        print("Koi expense nahi hai.")
        return
    for i, e in enumerate(expenses, 1):
        print(f"{i}. {e['date']} | {e['category']} | {e['amount']} | {e['note']}")


def show_incomes():
    print("\n--- Apki Incomes ---")
    if not incomes:
        print("Koi income nahi hai.")
        return
    for i, inc in enumerate(incomes, 1):
        print(f"{i}. {inc['date']} | {inc['source']} | {inc['amount']} | {inc['note']}")


def dashboard():
    total_income  = sum(i["amount"] for i in incomes)
    total_expense = sum(e["amount"] for e in expenses)
    balance       = total_income - total_expense

    print("\n====== Dashboard ======")
    print(f"Total Income  : {total_income}")
    print(f"Total Expense : {total_expense}")
    print(f"Balance       : {balance}")
    print("=======================")


def delete_expense():
    show_expenses()
    if not expenses:
        return
    num = int(input("Konsa number delete karna hai? ")) - 1
    removed = expenses.pop(num)
    print(f"Delete ho gaya: {removed['category']} - {removed['amount']}")


def main():
    print("============================")
    print("   FinTrack - Simple Python ")
    print("============================")

    while True:
        print("\n1. Expense Add Karo")
        print("2. Income Add Karo")
        print("3. Expenses Dekho")
        print("4. Incomes Dekho")
        print("5. Dashboard")
        print("6. Expense Delete Karo")
        print("7. Exit")

        choice = input("\nOption chuno: ")

        if   choice == "1": add_expense()
        elif choice == "2": add_income()
        elif choice == "3": show_expenses()
        elif choice == "4": show_incomes()
        elif choice == "5": dashboard()
        elif choice == "6": delete_expense()
        elif choice == "7":
            print("Allah Hafiz!")
            break
        else:
            print("Galat option, dobara try karo.")


main()
```
